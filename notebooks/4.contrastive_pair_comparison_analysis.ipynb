```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
raw_gt = pd.read_csv('/content/raw_ground_truth.csv')
ref_gt = pd.read_csv('/content/reference_ground_truth.csv')

# Display the first few rows of the datasets to understand their structure
display(raw_gt, ref_gt)
```

```python
import seaborn as sns

def plot_distributions(raw_data, ref_data, attributes, title, label1, label2, title_size=25):

    # Calculate the average of the selected attributes for both raw and reference data
    raw_avg = raw_data[attributes].mean(axis=1)
    ref_avg = ref_data[attributes].mean(axis=1)

    plt.figure(figsize=(12, 8))

    sns.histplot(raw_avg, kde=True, color='skyblue', bins=30, label = label1)
    sns.histplot(ref_avg, kde=True, color = 'lightpink', bins=30, label = label2)
    plt.title(title, fontsize=title_size)
    plt.ylabel('frequency')
    plt.legend(fontsize = title_size)

    plt.show()
```

```python

display(raw_gt.columns)
display(ref_gt.columns)
```

```python
Colorfulness = ['Vibrant and vivid']
plot_distributions(ref_gt, raw_gt, Colorfulness, 'Contrastive Loss Analysis: Raw VS Reference', 'raw', 'reference')
White_balance = ['Accurate Color Representation']
plot_distributions(ref_gt, raw_gt, White_balance, 'White Balance Evaluation: Raw VS Reference', 'raw', 'reference')
Exposure = ['Crystal-clear', 'Well-Lit Underwater', 'Balanced Exposure']
plot_distributions(ref_gt, raw_gt, Exposure, 'Lighting Quality Assessment: Raw VS Reference', 'raw', 'reference')
```

```python
base_result = pd.read_csv('/content/base_results.csv')
raw_result = pd.read_csv('/content/RAW_results.csv')
vivid_result = pd.read_csv('/content/vivid_mid_results.csv')
color_result = pd.read_csv('/content/color_cast_results.csv')
lit_result = pd.read_csv('/content/exposure_results.csv')


display(base_result, raw_result, vivid_result, color_result, lit_result)
```

```python
base_result.columns
WB = ['Accurate Color Representation']
CL = ['Vibrant and vivid', 'Bright, Colorful Underwater Scene']
LIT = ['Crystal-clear and Unobstructed Scene', 'Balanced and Well-Lit view']
```

```python
plot_distributions(vivid_result, base_result, CL, 'Contrastive Loss Analysis: Enhanced Model vs. Base Model', 'enhanced model', 'base model')
plot_distributions(color_result, base_result, WB, 'White Balance Evaluation: Enhanced Model vs. Base Model', 'enhanced model','base model')
plot_distributions(lit_result, base_result, LIT, 'Lighting Quality Assessment: Enhanced Model vs. Base Model', 'enhanced model', 'base model')
```

```python
def calculate_case_level_stats(df1, df2, attributes):
    # Compute the difference at the case level
    case_diff = df1[attributes] - df2[attributes]

    # Calculate the mean and std of the differences
    mean_diff = case_diff.mean()
    std_diff = case_diff.std()

    return mean_diff, std_diff

def print_stats(mean_diff, std_diff, comparison_type, attribute_type):
    print(f"Mean and Std for {comparison_type} on {attribute_type}: {mean_diff}, {std_diff}")

attribute_groups = {
    'CL': CL,
    'WB': WB,
    'LIT': LIT,
    'Colorfulness': Colorfulness,
    'White balance': White_balance,
    'Exposure': Exposure
}

# Compute and print stats for Reference vs Raw
for attr_type, attrs in attribute_groups.items():
   if attr_type in ['Colorfulness', 'White balance', 'Exposure']:
    if attr_type == 'Colorfulness':
      mean_diff, std_diff = calculate_case_level_stats(ref_gt, raw_gt, attrs)
    elif attr_type == 'White balance':
      mean_diff, std_diff = calculate_case_level_stats(ref_gt, raw_gt, attrs)
    elif attr_type == 'White balance':
      mean_diff, std_diff = calculate_case_level_stats(ref_gt, raw_gt, attrs)
    print_stats(mean_diff, std_diff, 'Reference vs Raw', attr_type)

# Compute and print stats for Enhanced vs Base
for attr_type, attrs in attribute_groups.items():
    if attr_type in ['CL', 'WB', 'LIT']:  # Assuming vivid_result, color_result, lit_result apply to these only
        if attr_type == 'CL':
            mean_diff, std_diff = calculate_case_level_stats(vivid_result, base_result, attrs)
        elif attr_type == 'WB':
            mean_diff, std_diff = calculate_case_level_stats(color_result, base_result, attrs)
        elif attr_type == 'LIT':
            mean_diff, std_diff = calculate_case_level_stats(lit_result, base_result, attrs)
        print_stats(mean_diff, std_diff, 'Enhanced vs Base', attr_type)
```

```python
def append_to_results_df(mean_diff, std_diff, comparison_type, attribute_type, results_df):
    for attr in mean_diff.index:
        results_df = results_df.append({
            "Comparison Type": comparison_type,
            "Attribute Type": attribute_type,
            "Attribute": attr,
            "Mean Difference": mean_diff[attr],
            "Std Deviation": std_diff[attr]
        }, ignore_index=True)
    return results_df


results_df = pd.DataFrame(columns=["Comparison Type", "Attribute Type", "Attribute", "Mean Difference", "Std Deviation"])


for attr_type, attrs in attribute_groups.items():
    if attr_type in ['Colorfulness', 'White balance', 'Exposure']:
        mean_diff, std_diff = calculate_case_level_stats(ref_gt, raw_gt, attrs)
        results_df = append_to_results_df(mean_diff, std_diff, 'Reference vs Raw', attr_type, results_df)

for attr_type, attrs in attribute_groups.items():
    if attr_type in ['CL', 'WB', 'LIT']:
      if attr_type == 'CL':
        mean_diff, std_diff = calculate_case_level_stats(vivid_result, base_result, attrs)
      elif attr_type == 'WB':
        mean_diff, std_diff = calculate_case_level_stats(color_result, base_result, attrs)
      elif attr_type == 'LIT':
        mean_diff, std_diff = calculate_case_level_stats(lit_result, base_result, attrs)


      results_df = append_to_results_df(mean_diff, std_diff, 'Enhanced vs Base', attr_type, results_df)


display(results_df)
```

```python
cl_df = results_df[results_df['Attribute Type'] == 'CL']
wb_df = results_df[results_df['Attribute Type'] == 'WB']
lit_df = results_df[results_df['Attribute Type'] == 'LIT']
exp_df = results_df[results_df['Attribute Type'] == 'Exposure']

combined_exp_mean = exp_df['Mean Difference'].mean()
combined_exp_std = exp_df['Std Deviation'].mean()

combined_exp_row = {
    'Comparison Type': 'Reference vs Raw',
    'Attribute Type': 'Exposure',
    'Attribute': 'Crystal-clear; Well-Lit Underwater; Balanced Exposure	',
    'Mean Difference': combined_exp_mean,
    'Std Deviation': combined_exp_std
}
# Now, calculate the mean of the mean differences and the mean of the standard deviations
combined_cl_mean = cl_df['Mean Difference'].mean()
combined_cl_std = cl_df['Std Deviation'].mean()
combined_wb_mean = wb_df['Mean Difference'].mean()
combined_wb_std = wb_df['Std Deviation'].mean()
combined_lit_mean = lit_df['Mean Difference'].mean()
combined_lit_std = lit_df['Std Deviation'].mean()
# Add a new row with these combined values
combined_cl_row = {
    'Comparison Type': 'Enhanced vs Base',
    'Attribute Type': 'CL',
    'Attribute': 'Vibrant and vivid; Bright, Colorful Underwater Scene',
    'Mean Difference': combined_cl_mean,
    'Std Deviation': combined_cl_std
}
combined_wb_row = {
    'Comparison Type': 'Enhanced vs Base',
    'Attribute Type': 'WB',
    'Attribute': 'Accurate Color Representation',
    'Mean Difference': combined_wb_mean,
    'Std Deviation': combined_wb_std
}
combined_lit_row = {
    'Comparison Type': 'Enhanced vs Base',
    'Attribute Type': 'LIT',
    'Attribute': 'Crystal-clear and Unobstructed Scene; Balanced and Well-Lit view',
    'Mean Difference': combined_lit_mean,
    'Std Deviation': combined_lit_std
}
results_df = results_df.append(combined_exp_row, ignore_index=True)

results_df = results_df.append(combined_cl_row, ignore_index=True)
results_df = results_df.append(combined_wb_row, ignore_index = True)
results_df = results_df.append(combined_lit_row, ignore_index = True)

results_df
```
